```python
import pandas as pd
import seaborn as sns
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt

iowa_file_path = 'data-input/home-ml/train.csv'
iowa_data_predict_path = 'data-input/home-ml/test.csv'

home_data = pd.read_csv(iowa_file_path)
home_data_predict = pd.read_csv(iowa_data_predict_path)
# Create target object and call it y
y = home_data.SalePrice
# Create X
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = home_data[features]
# Split into validation and training data
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)

def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0) #define modelo
    model.fit(train_X, train_y) #treina(fit) o modelo
    preds_val = model.predict(val_X) #prediz valores
    mae = mean_absolute_error(val_y, preds_val) #obtem métrica de Média Absoluta de Erros nas predições
    return(mae)
```

```python
list_mae = []
list_leaf_nodes = []

for max_leaf_nodes in range(67,74): #for max_leaf_nodes in [5, 50, 500, 5000]
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    list_leaf_nodes.append(max_leaf_nodes)
    list_mae.append(my_mae)
    #print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
    
#mae_df = pd.DataFrame(list(zip(list_leaf_nodes,list_mae)),columns = ['leaf nodes','mae'])

#print(mae_df.describe())
#print(mae_df.head())
plt.plot(list_leaf_nodes,list_mae)
plt.show()
```

```python
def pandas_scatter(df,column1,column2):
    #sns.set_theme()
    
    #sns.scatterplot(data=df,x=column1,y=column2)
    plt.scatter(df[column1],df[column2])
    plt.xlabel(f'{column1}')
    plt.ylabel(f'{column2}')
    plt.show()
    

def all_columns_scatter(df):
    for column1 in df.columns:
        for column2 in df.columns:
            if column1!='Id' and column2!='Id':
                if (df.dtypes[column1]=='int64' or df.dtypes[column1]=='float64') and (df.dtypes[column2]=='int64' or df.dtypes[column2]=='float64'):
                    if column1 != column2:
                        #print(column1,column2,df.dtypes[column1],df.dtypes[column2])
                        pandas_scatter(df,column1,column2)
    

def target_column_scatter(df,target):
    for column in df.columns:
        if column!='Id' and column!=target:
            if (df.dtypes[column]=='int64' or df.dtypes[column]=='float64'):
                pandas_scatter(df,column,target)
```

# Testando com modelo RandomForestGenerator e features optimizadas

## O quão efetivo o modelo será?

```python
iowa_file_path = 'data-input/home-ml/train.csv'
iowa_data_predict_path = 'data-input/home-ml/test.csv'

home_data = pd.read_csv(iowa_file_path).fillna(method='pad')
home_data_predict = pd.read_csv(iowa_data_predict_path).fillna(method='pad')

y = home_data.SalePrice
# Create X
features = ['OverallQual', 'TotalBsmtSF', '1stFlrSF', 'GrLivArea', 'GarageCars', 'GarageArea']
X = home_data[features]
# Split into validation and training data
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)

rfmodel = RandomForestClassifier(random_state=0)

rfmodel.fit(train_X,train_y)
preds_val = rfmodel.predict(val_X)

mae = mean_absolute_error(val_y, preds_val) #obtem métrica de Média Absoluta de Erros nas predições

print(mae)
```

```python
def get_mae_rf(train_X, val_X, train_y, val_y,n_estimators=100,max_depth=None,max_leaf_nodes=None): #função para calcular mae, variando parâmetros
    rfmodel = RandomForestClassifier(n_estimators=n_estimators,max_depth=max_depth,max_leaf_nodes=max_leaf_nodes,random_state=0)

    rfmodel.fit(train_X,train_y)
    preds_val = rfmodel.predict(val_X)

    mae = mean_absolute_error(val_y, preds_val)
    return mae
    
```

```python
mae = get_mae_rf(train_X, val_X, train_y, val_y) #teste se função está em valores default
print(mae)
```

```python
for n_estimators in [10,50,100,150,200,250,500]: #testando melhor valor de n_estimators
    mae = get_mae_rf(train_X, val_X, train_y, val_y,n_estimators=n_estimators)
    print(n_estimators,mae)
```

```python
for max_depth in [10,50,100,150,200,250,500]: #testando melhor valor de max_depth
    mae = get_mae_rf(train_X, val_X, train_y, val_y,max_depth=max_depth)
    print(max_depth,mae)
```

```python
for max_leaf_nodes in [10,50,100,150,200,250,500]: #testando melhor valor de max_leaf_nodes
    mae = get_mae_rf(train_X, val_X, train_y, val_y,max_leaf_nodes=max_leaf_nodes)
    print(max_leaf_nodes,mae)
```

```python
#agregado de variações:
val_mae_rf_list = []
for n_estimators in [10,50,100,150,200,250,500]:
    for max_depth in [10,50,100,150,200,250,500]:
        for max_leaf_nodes in [10,50,100,150,200,250,500]:
            mae = get_mae_rf(train_X, val_X, train_y, val_y,n_estimators=n_estimators,max_depth=max_depth,max_leaf_nodes=max_leaf_nodes)
            val_mae_rf_list.append(mae)
            print(f'n_estimators = {n_estimators} | max_depth = {max_depth} | max_leaf_nodes = {max_leaf_nodes} | mae = {mae}')

print(val_mae_rf_list)
print(f'Menor valor é: {min(val_mae_rf_list)}')
```

```python
#menor valor: n_estimators = 50 | max_depth = Irrelevante | max_leaf_nodes = 250 - menor valor se repete nestas condições
#mae deve ser = 26589.320547945204, se max_depth for = None e os demais parametros forem iguais ao comentário acima
mae = get_mae_rf(train_X, val_X, train_y, val_y,n_estimators=50,max_depth=None,max_leaf_nodes=250)
print(mae)
```

# Testando threading para rodar os modelos mais rápidamente
## OBS: não rodar este bloco de código, é necessário limitar o número de threads para que ele não crashe

```python
# OBS: não rodar este bloco de código, é necessário limitar o número de threads para que ele não crashe
from threading import Thread
from time import perf_counter


def rf_optimal_pars(train_X, val_X, train_y, val_y,n_estimators,max_depth,max_leaf_nodes): #função a ser executada pelas threads
    mae = get_mae_rf(train_X, val_X, train_y, val_y,n_estimators=n_estimators,max_depth=max_depth,max_leaf_nodes=max_leaf_nodes)
    print(f'n_estimators = {n_estimators} | max_depth = {max_depth} | max_leaf_nodes = {max_leaf_nodes} | mae = {mae}')

# crias as threads
threads = []
for n_estimators in [10,50,100,150,200,250,500]:
        for max_depth in [10,50,100,150,200,250,500]:
            for max_leaf_nodes in [10,50,100,150,200,250,500]:
                t = Thread(target=rf_optimal_pars, args=(train_X, val_X, train_y, val_y,n_estimators,max_depth,max_leaf_nodes))
                threads.append(t)

# inicia as threads
for thread in threads:
    thread.start()

# espera as threads serem completadas
for thread in threads:
    thread.join()

```
